```python
import pandas as pd

df_fertilizantes = pd.read_csv("Fertilizantes Sniim.csv")
df_fertilizantes.head()
```

```python
df_fertilizantes.shape
```

```python
df_fertilizantes.info()
```

```python
df_fertilizantes["Fecha"] = pd.to_datetime(
    df_fertilizantes["Fecha"],
    dayfirst=True,
    errors="coerce"
)
```

```python
df_fertilizantes.dtypes
```

```python
columnas_precio = ["Precio Minimo", "Precio Maximo", "Precio Frecuente"]
for col in columnas_precio:
    df_fertilizantes[col] = (
        df_fertilizantes[col]
        .astype(str)
        .str.replace(",", "", regex=False)
        .str.strip()
    )

    df_fertilizantes[col] = pd.to_numeric(
        df_fertilizantes[col],
        errors="coerce"
    )
df_fertilizantes.dtypes
```

```python
df_fertilizantes.isnull().sum()
```

```python
df_fertilizantes.head()
```

```python
df_fertilizantes["Producto"].unique()
```

```python
df_fertilizantes["Producto"].value_counts()
```

```python
df_fertilizantes["Concentración (N-P-K)"].unique()
```

```python
df_fertilizantes[["N", "P", "K"]] = df_fertilizantes[
    "Concentración (N-P-K)"
].str.split("-", expand=True)
df_fertilizantes[["N", "P", "K"]] = df_fertilizantes[
    ["N", "P", "K"]
].astype(float)
```

```python
df_fertilizantes.head()
```

```python
cols = list(df_fertilizantes.columns)

cols.remove("N")
cols.remove("P")
cols.remove("K")

posicion = cols.index("Concentración (N-P-K)") + 1

cols = (
    cols[:posicion] +
    ["N", "P", "K"] +
    cols[posicion:]
)

df_fertilizantes = df_fertilizantes[cols]
```
