```python
import pandas as pd
import numpy as np

df = pd.read_csv("Tipos de suelos-d1403_reverso.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
#Eliminar columnas
columnas_eliminar = [
    "Relieve",
    "% de arcillas",
    "% de limo",
    "% de arena"
]

df_limpio = df.drop(columns=columnas_eliminar)
df_limpio.head()
```

```python
df_limpio.isnull().sum()
```

```python
df_limpio["faltantes_por_fila"] = df_limpio.isnull().sum(axis=1)
df_limpio[["faltantes_por_fila"]].head(20)
```

```python
df_limpio[df_limpio["faltantes_por_fila"] > 5]
```

```python
df_limpio = df_limpio[df_limpio["faltantes_por_fila"] <= 5]
df_limpio = df_limpio.drop(columns=["faltantes_por_fila"])
```

```python
df_limpio
```

```python
# Reiniciar índice
df_limpio = df_limpio.reset_index(drop=True)

# Ver resultado
df_limpio.head()
```
